# *Plot's for Continuous Data*

```python
#Histogram/Distribution Plot
```

```python
sns.displot(df['total_bill'])
plt.show()
```

```python
sns.displot(df['total_bill'],bins=10,kde=True)
plt.title("Histogram of total_bill")
plt.show()
```

# Box Plot

```python
sns.boxplot(x=df['total_bill'])
plt.show()
```

# Scatter Plot

```python
sns.scatterplot(x=df['total_bill'],y=df['tip'])
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,size='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker',style='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker',col='size')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker',col='size')
plt.show()
```

# Line Plot

```python
df['sno']=pd.DataFrame(np.arange(1245))
df
```

```python
sns.relplot(x='sno',y='tip',kind='line',data=df)
plt.show()
```

```python
df.drop("sno",axis=1,inplace=True)
```

# JoinPlot

```python
sns.jointplot(y='tip',x='total_bill',data=df)
plt.show()
```

# Vlolin Plot

```python
sns.violinplot(x="total_bill",y="day",data=df)
plt.show()
```

```python
sns.violinplot(x="total_bill",y="day",data=df,palette='rainbow')
plt.show()
```

# Pairplot for multiple continous variable

```python
sns.pairplot(df)
plt.show()
```

```python
continous=["total_bill",'tip']
sns.pairplot(df,vars=continous)
plt.show()
```

```python
continous=["total_bill",'tip']
sns.pairplot(df,vars=continous,hue='sex')
plt.show()
```

# Heat Map

```python
corr_matrix=df[['total_bill','tip']].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

# Bar Plot for single variable

```python
sns.catplot(x='sex',y='total_bill',data=df)
plt.show()
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
plt.show()
```

```python
df['size'].value_counts()
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind="bar",hue='size')
plt.show()
```

```python
sns.catplot(x='smoker', y='tip',data=df,kind="bar",order=['Yes','No'],hue='size')
plt.show()
```
